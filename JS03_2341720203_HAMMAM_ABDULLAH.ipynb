**Separate usable variables from unusable variables.**

```python
X = df1.drop(columns=["id", "Unnamed: 32", "diagnosis"], errors="ignore")

corr_matrix = X.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col] > 0.8)]
X = X.drop(columns=to_drop)
print("Dropped (high correlation):", to_drop)
```

I removed uninformative columns such as id and Unnamed: 32, keeping only relevant numeric medical features for further modeling.

**Encode the "diagnosis" column.**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(df1["diagnosis"])
```

The diagnosis column was encoded using LabelEncoder, converting categorical labels into numeric form, where 0 represents benign and 1 malignant.

**Standardize all columns with numeric values.**

```python
num_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
num_tf = Pipeline([("scaler", StandardScaler())])
print(X.head())
```

I applied StandardScaler to normalize numeric columns, ensuring all features share the same scale for consistent and stable model training.

**Perform feature selection using SelectKBest.**

```python
from sklearn.feature_selection import f_classif

selector_filter = SelectKBest(score_func=f_classif, k=9)


num_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()


preprocess = ColumnTransformer([
    ("num", num_tf, num_cols),
], remainder="drop")

pipe_filter = Pipeline([
    ("prep", preprocess),
    ("sel", selector_filter),
    ("clf", LogisticRegression(max_iter=1000))
])
```

Feature selection was done with SelectKBest using ANOVA F-test, keeping only the most informative predictors to reduce redundancy and noise.

**Perform the testing process using the Logistic Regression model as in Lab 1.**

**You can use a model pipeline to simplify your work..**


```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

pipe_filter.fit(X_train, y_train)
pred_train = pipe_filter.predict(X_train)
print("=== Filter (ANOVA) + LR ===")
print("Training Accuracy:", accuracy_score(y_train, pred_train))
print(classification_report(y_train, pred_train))
pred = pipe_filter.predict(X_test)
print("=== Filter (ANOVA) + LR ===")
print("Testing Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

I trained Logistic Regression on the processed dataset, achieving around 91% accuracy with balanced performance across precision, recall, and F1-score.

```python
feat_names = pipe_filter.named_steps["prep"].get_feature_names_out()
sel = pipe_filter.named_steps["sel"]

mask = sel.get_support()
selected_names = feat_names[mask]
selected_scores = sel.scores_[mask]

top = sorted(zip(selected_names, selected_scores), key=lambda t: t[1], reverse=True)[:10]
print("Top fitur:")
for item in top:
    print(item)
```
